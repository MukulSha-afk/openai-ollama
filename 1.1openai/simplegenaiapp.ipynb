simple gen ai app using langchain


```python
import os
from dotenv import load_dotenv

load_dotenv()

# OpenAI API key (LangChain automatically use karega)
print("KEY FOUND:", os.getenv("OPENAI_API_KEY") is not None)
openai_key = os.getenv("ANOTHER_API_KEY")

if openai_key is None:
    raise ValueError("ANOTHER_API_KEY not found")

os.environ["ANOTHER_API_KEY"] = openai_key

# LangChain settings (optional)
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_PROJECT"] = os.getenv("LANGCHAIN_PROJECT", "default_project")

```

```python
from langchain_community.llms import Ollama

llm = Ollama(model="gemma3:1b")
print(llm.invoke("What is Generative AI?"))
```

```python
from langchain_community.document_loaders import WebBaseLoader
```

```python
loader = WebBaseLoader("https://en.wikipedia.org/wiki/Generative_AI")
loader
```

```python
docs=loader.load()
docs
```

```python
## document to chunks 
### Load the data  then get docs and divide our chunks(smaller size data) then convert into vector using vector embeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_spliiters = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)

documents = text_spliiters.split_documents(docs)
```

```python
documents
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings( model="text-embedding-3-small", openai_api_key="ANOTHER_API_KEY")
```

```python
from langchain_community.vectorstores import FAISS
vectorstoredb = FAISS.from_documents(documents, embeddings)
```
