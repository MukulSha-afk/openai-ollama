```python
import os
from dotenv import load_dotenv

load_dotenv()

# OpenAI API key (LangChain automatically use karega)
openai_key = os.getenv("ANOTHER_API_KEY")

if openai_key is None:
    raise ValueError("ANOTHER_API_KEY not found")

os.environ["ANOTHER_API_KEY"] = openai_key

# LangChain settings (optional)
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_PROJECT"] = os.getenv("LANGCHAIN_PROJECT", "default_project")

```

```python
from langchain_community.llms import Ollama

llm = Ollama(model="gemma3:1b")
print(llm.invoke("What is Generative AI?"))
```

```python
## Input and get responce from LLM
```

```python
### new # --- IGNORE ---
#Chatpromt
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate
prompt = ChatPromptTemplate.from_messages([
    ("system", "you are a helpful assistant that helps people find information."),
    ("human", "{input}")
]
                                          
                                          )
prompt
```

```python
## chain- combine the prompt and llm
chain = prompt | llm

responce=chain.invoke({"input": "Explain the theory of relativity in simple terms."})
responce
```

```python
type(responce)
```

```python
## stroutput Parser
from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
chain = prompt | llm | output_parser

responce = chain.invoke({"input": "Explain the theory of relativity in simple terms."})
responce
```
